## **IV. Modelisation<a class="anchor" id="m"></a>**

To start, we need to clarify the core need of the problem, which will guide our objectives and constraints.

**Need Identification:** A logistics provider needs to deliver goods to multiple customers efficiently. Efficient delivery means achieving optimal routes for multiple delivery vehicles to meet customer requirements while minimizing operational costs.

**Problem Type:** This problem can be formulated as a Vehicle Routing Problem (VRP) if multiple vehicles are used. If only one vehicle is used, it aligns with the Traveling Salesman Problem (TSP). In our case it's VRP.

### **1. Comprehensive Problem Representation<a class="anchor" id="cp"></a>**

#### **A. Why VRP and Not TSP?<a class="anchor" id="wp"></a>**
This problem is categorized as a **Vehicle Routing Problem (VRP)** rather than a **Traveling Salesman Problem (TSP)** because:  
- In the TSP, a single vehicle must visit all customers and return to the depot.  
- In contrast, our scenario involves **multiple vehicles**, requiring the optimization of customer allocation to vehicles and the determination of the best route for each vehicle.  

The VRP adds complexity compared to TSP, as it must address the assignment of customers to vehicles, while also considering factors like vehicle capacity and time windows.

##### **<u>Key Constraints</u>**

1. **Single Visit per Customer**
   - **Description:** Each customer must be visited exactly once, and only one vehicle can serve each customer.  
   - **Constraint Type:** Equality constraint.  
   - **Purpose:** Ensures that no customer is missed or served multiple times, avoiding redundant or overlapping deliveries.

2. **Depot Start and End**
   - **Description:** All vehicles must begin and end their routes at the depot (node 0).  
   - **Constraint Type:** Equality constraint.  
   - **Purpose:** Ensures proper route closure for operational consistency and planning.

3. **Capacity Constraint**
   - **Description:** Each vehicle has a maximum carrying capacity that cannot be exceeded by the total demand of its assigned customers.  
   - **Constraint Type:** Hard constraint.  
   - **Purpose:** Prevents vehicle overloading and ensures feasible route assignments.

4. **Time Windows**
   - **Description:** Deliveries must be made within a specified time window for each customer.  
   - **Constraint Type:** Hard or soft constraint (with penalties for violations).  
   - **Purpose:** Improves customer satisfaction and ensures realistic delivery schedules.

5. **Service Time**
   - **Description:** Time spent at each customer location for unloading or completing the service is included in the schedule.  
   - **Constraint Type:** Operational constraint.  
   - **Purpose:** Ensures realistic timing for routing and adherence to delivery schedules.

6. **Delivery-Leave Continuity**
   - **Description:** After serving a customer, a vehicle must leave either to visit another customer or to return to the depot.  
   - **Constraint Type:** Flow conservation constraint.  
   - **Purpose:** Ensures route continuity and prevents incomplete or disjointed paths.

#### **B. Mathematic Modelisation<a class="anchor" id="mm"></a>**
In this section, we present the mathematical modeling of the Vehicle Routing Problem (VRP), defining the necessary constraints and objectives for optimizing the routes. Each constraint is formulated using decision variables that model the relationships between customers, vehicles, and the depot, with the goal of minimizing the total cost while adhering to specific requirements such as capacity, time windows, and service times. Additionally, we define a fitness function, which serves to evaluate and optimize the solution, ensuring that the routes are both efficient and feasible according to the problem's constraints.


##### **<u>Fitness Function for a Simple VRP</u>**


In a basic **Vehicle Routing Problem (VRP)**, the primary objective is to minimize the total distance traveled by all vehicles while ensuring that all customers are visited exactly once. The fitness function can be expressed as:

$$
\text{Minimize } Z_{\text{VRP}} = \sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \cdot x_{ij}^k
$$

**Where:**
- $c_{ij}$: The cost (e.g., distance or time) of traveling from node $i$ to node $j$.
- $x_{ij}^k$: A binary variable that equals 1 if vehicle $k$ travels from $i$ to $j$, and 0 otherwise.
- $V$: The set of all nodes (customers and depot).
- $K$: The set of available vehicles.

This fitness function simply minimizes the total travel cost, without considering additional constraints like vehicle capacity or customer-specific time windows.


##### **<u>From VRP to CVRPTW</u>**

While the simple VRP aims to minimize travel costs, real-world routing problems often include additional constraints, such as vehicle capacity and customer time windows. These extensions lead to the **Capacitated Vehicle Routing Problem with Time Windows (CVRPTW)**. The CVRPTW adds complexity by requiring that:

1. **Vehicle Capacities:** Each vehicle has a limited carrying capacity, and the total demand of customers assigned to a vehicle cannot exceed this capacity.
2. **Time Windows:** Each customer has a specific time window during which they must be serviced.

To incorporate these constraints, the fitness function must account for **multiple objectives**, such as minimizing the total cost while adhering to these constraints. Additionally, the model must ensure feasible solutions by enforcing capacity and time window constraints.



##### **<u>Fitness Function for the CVRPTW</u>**

The fitness function for the CVRPTW combines the objectives of minimizing the number of vehicles and the total travel cost, while ensuring feasible delivery times and capacity usage. A common formulation is:

$$
\text{Minimize } Z_{\text{CVRPTW}} = \left( \sum_{k \in K} T_k \right) \cdot \left( \sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \cdot x_{ij}^k + \sum_{i \in V} s_i \right)
$$

**Where:**
- $\sum_{k \in K} T_k$: Total number of vehicles used.
- $\sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \cdot x_{ij}^k$: Total travel cost (e.g., distance, time, or fuel).
- $\sum_{i \in V} s_i$: Total service time for all customers.
- $x_{ij}^k$: Binary decision variable indicating whether vehicle $k$ travels from customer $i$ to customer $j$.
- $T_k$: Binary decision variable indicating whether vehicle $k$ is used.


##### **<u>Why Extend to the CVRPTW?</u>**

1. **Realistic Constraints:** In real-world applications, simple VRP models fail to capture operational challenges like limited vehicle capacity and customer time preferences. Adding these constraints makes the model more realistic and applicable to industries like logistics and supply chain management.
   
2. **Minimizing Total Costs:** By multiplying the number of vehicles used by the total travel cost, this fitness function penalizes solutions that use more vehicles or inefficient routing. It ensures a balanced trade-off between reducing the fleet size and optimizing routes.

3. **Feasibility:** Enforcing time windows and capacity ensures that the solution is not only optimal in terms of cost but also operationally feasible.

4. **Multi-Objective Nature:** The CVRPTW inherently involves multiple objectives (e.g., minimizing distance, adhering to time windows, and minimizing the fleet size). This fitness function integrates these objectives into a unified mathematical framework.


While the simple VRP focuses on minimizing travel costs, the CVRPTW addresses more complex, real-world challenges by incorporating capacity constraints and time windows. The fitness function evolves to balance multiple objectives, ensuring feasible and cost-effective routing solutions. By integrating additional constraints into the fitness function, the CVRPTW model becomes a powerful tool for solving practical logistics and delivery problems.



##### **<u>Constraints</u>**

##### **1. <u>Single Visit per Customer Constraint</u>**

**Description:**  
Each customer must be visited exactly once by a single vehicle. No customer can be omitted or visited multiple times.

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.

**Model:**  
$$
\sum_{k=1}^K \sum_{j=1, j \neq i}^N x_{ij}^k = 1 \quad \forall i \in \{1, \dots, N\}
$$  
This ensures that each customer is visited exactly once.

**Explanation:**  
This constraint ensures that each customer is visited only once by a vehicle, and no customer will be missed or visited multiple times.

---

##### **2. <u>Depot Start and End Constraint</u>**

**Description:**  
Each vehicle must start and end its route at the depot (node 0).

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.

**Model:**  
$$
\sum_{j=1}^N x_{0j}^k = 1 \quad \forall k \in \{1, \dots, K\} \quad \text{(Vehicle starts from the depot)}
$$  
$$
\sum_{i=1}^N x_{i0}^k = 1 \quad \forall k \in \{1, \dots, K\} \quad \text{(Vehicle returns to the depot)}
$$

**Explanation:**  
This constraint ensures that each vehicle starts and ends its route at the depot, thus completing the delivery loop.

---

##### **3. <u>Delivery-Leave Continuity</u>**

**Description:**  
When a vehicle arrives at a customer, it must either proceed to another customer or return to the depot. This ensures the continuity of the route without any breaks.

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.  
- $t_i^k$: Arrival time at customer $i$ for vehicle $k$.

**Model:**  
1. **Inflow and Outflow for each customer $i$ and vehicle $k$:**
   $$
   \sum_{j \neq i} x_{ij}^k = \sum_{j \neq i} x_{ji}^k \quad \forall i \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
   $$  
   This ensures that a vehicle arrives at each customer exactly once and departs exactly once.

2. **Arrival time calculation $t_i^k$:**
   $$
   t_j^k \geq t_i^k + s_i + d_{ij} - M(1 - x_{ij}^k) \quad \forall i, j \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
   $$  
   Where $M$ is a large constant to ensure the constraint is inactive if $x_{ij}^k = 0$.

**Explanation:**  
This constraint ensures that vehicles continue their routes without interruption and that the arrival time at each customer is calculated based on the previous route.

---

##### **4. <u>Time Window Constraint</u>**

**Description:**  
Each customer has a specific time window during which they wish to receive their delivery. The delivery must occur within this window, or with a penalty if the constraint is softened.

**Decision Variable:**  
- $t_i^k$: Arrival time at customer $i$ for vehicle $k$.

**Model:**  
$$
a_i \leq t_i^k \leq b_i \quad \forall i \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
$$  
Where $a_i$ and $b_i$ are the time window bounds for customer $i$.

**Explanation:**  
This constraint ensures that each delivery respects the time window defined for each customer. If a time window is exceeded, a penalty could be added to the model.

---

##### **5. <u>Capacity and Demand Constraint</u>**

**Description:**  
Each vehicle has a maximum capacity, and each customer has a specific demand that must be met without exceeding this capacity.

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.  
- $q_k$: Capacity of vehicle $k$.  
- $d_i$: Demand of customer $i$.

**Model:**  
$$
\sum_{i=1}^N \sum_{j=1, j \neq i}^N d_i \cdot x_{ij}^k \leq q_k \quad \forall k \in \{1, \dots, K\}
$$

**Explanation:**  
This constraint ensures that the demand of each customer is met while respecting the capacity of each vehicle. The sum of the demands for all customers visited by a vehicle must not exceed the capacity of the vehicle.

---

##### **6. <u>Service Time Constraint</u>**

**Description:**  
The service time represents the time required for a vehicle to complete a delivery at a customer, including unloading time or waiting for acceptance.

**Decision Variable:**  
- $t_i^k$: Arrival time at customer $i$ for vehicle $k$.  
- $s_i$: Service time at customer $i$.

**Model:**  
$$
t_j^k \geq t_i^k + s_i + d_{ij} - M(1 - x_{ij}^k) \quad \forall i, j \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
$$

**Explanation:**  
This constraint accounts for the service time at each customer location, ensuring that the scheduling of routes and timings is realistic.

---

### **2. Metaheuristic Techniques and Formal Modelling Detail<a class="anchor" id="mt"></a>**

#### **A. Complexity<a class="anchor" id="cx"></a>**

##### **<u>VRPTW and its Relation to VRP</u>**

VRPTW is an extension of VRP. In VRPTW, each customer has a specific time window during which they must be visited, in addition to vehicle capacity constraints.  
To reduce VRPTW to VRP, we can consider a simplified case where time windows are the same as customer schedules, so they impose no constraints on delivery times. This simplification removes the time window constraints, reducing VRPTW to a standard case of VRP. Consequently, any algorithm that solves VRPTW can also solve VRP as a special case, indicating that VRPTW is at least as difficult as VRP.

##### **<u>Proof that VRP is NP-Complete</u>**

##### **1. <u>VRP is in NP</u>**

To establish that the Vehicle Routing Problem (VRP) is in the class NP, we need to demonstrate that, given a candidate solution, we can verify its correctness in polynomial time. Specifically, we consider the Decision Version of VRP, where we ask whether there exists a set of routes for the vehicles such that all customers are served, all constraints are satisfied, and the total cost does not exceed a given value \(k\).

Given a proposed set of routes for the vehicles, the verification process involves the following checks, each of which can be performed in polynomial time:

- **a. Each customer is visited exactly once**  
  - Verification: Ensure that every customer appears in exactly one route.  
  - Time Complexity: Linear time $O(n)$, where $n$ is the number of customers.

- **b. Vehicle Capacity Constraints are Respected**  
  - Verification: For each vehicle, sum the demands of the customers on its route and ensure it does not exceed the vehicle’s capacity.  
  - Time Complexity: Linear time $O(n)$, as it involves iterating through each route and summing demands.

- **c. Routes Start and End at the Depot**  
  - Verification: Check that each route begins and ends at the designated depot.  
  - Time Complexity: Linear time $O(m)$, where $m$ is the number of vehicles (typically $m \leq n$).

- **d. Total Cost Does Not Exceed $k$**  
  - Verification: Calculate the total cost (e.g., total distance) by summing the costs of all routes and verify it is less than or equal to $k$.  
  - Time Complexity: Linear time $O(n)$, as it involves summing the costs of all routes.

Since all these verifications can be performed in linear time relative to the size of the input, VRP is in NP.

##### **2. <u>Reduction from Hamiltonian Cycle to VRPTW</u>**

To prove that VRP is NP-Hard, we will reduce a known NP-Complete problem to VRP in polynomial time. We choose the Hamiltonian Cycle Problem for this reduction.

**Hamiltonian Cycle Problem**  
- Input: An undirected graph $G = (V, E)$.  
- Question: Does there exist a cycle that visits each vertex exactly once and returns to the starting vertex?


**1. Reduction Process:**

- **Instance Transformation:**  
  - Given: An undirected graph $G = (V, E)$ for the Hamiltonian Cycle problem.  
  - Construct: A VRP instance as follows:  
    - **Depot:** Choose an arbitrary vertex $D$ from $V$ as the depot.  
    - **Customers:** All other vertices $V' = V \setminus D$ are treated as customers, each with a demand of 1.  
    - **Vehicle Capacity:** Set the vehicle capacity $C = |V|$, ensuring a single vehicle can serve all customers.  
    - **Distances:** Define the distance between two vertices $u$ and $v$ as:  
    $$
    d(u, v) = 
    \begin{cases} 
      1 & \text{if } (u, v) \in E \\
      2 & \text{if } (u, v) \notin E 
    \end{cases}
    $$
    - **Total Cost $k$:** Set $k = |V|$.

**2. Equivalence of Solutions:**  
- If $G$ has a Hamiltonian Cycle:  
  The corresponding VRPTW instance can be solved by having the single vehicle follow the Hamiltonian cycle, resulting in a total cost of $|V|$ (since each edge in the cycle has a distance of 1).
- If the VRP instance has a solution with total cost $\leq k$:  
  The vehicle’s route must use only edges with distance 1 to keep the total cost at $|V|$. This implies the existence of a Hamiltonian Cycle in $G$.

**3. Polynomial-Time Transformation:**  
- The reduction involves creating a VRP instance with:  
  - $|V|$ customers.  
  - A complete graph for distances, requiring $O(|V|^2)$ time.  
  - Setting capacities and costs, which are linear operations.  
- Therefore, the reduction is performed in polynomial time.

#### **B. Metaheuristic Algorithm<a class="anchor" id="ma"></a>**

##### **<u>Introduction to Metaheuristic Algorithms</u>**

Metaheuristic algorithms are problem-solving methods designed to find approximate solutions to complex optimization problems, especially those that are NP-hard. These algorithms are useful when the solution space is large or non-linear, like in the **Vehicle Routing Problem with Time Windows (VRPTW)**. Metaheuristics explore vast search spaces to find good-enough solutions in a reasonable time.

They are inspired by natural processes and handle large, complex tasks efficiently. While they do not guarantee optimal solutions, they typically find close-to-optimal solutions quickly.

##### **<u>Key Features of Metaheuristic Algorithms:</u>**
- **Exploration and Exploitation**: Balances searching broad areas (exploration) and refining promising areas (exploitation).
- **Stochastic Nature**: Randomness helps avoid getting stuck in local optima.
- **Adaptability**: Can be customized for various problems by modifying components like solution representation and operators (e.g., crossover, mutation).
- **Flexibility**: Can solve a wide range of combinatorial optimization problems, including VRPTW.

After implementing various algorithms that depend on a wide range of parameters, it is crucial to understand the impact of each parameter on the performance of the algorithm. This is why conducting an experimental study becomes a key step in evaluating and understanding the influence of each parameter.

In our case, we performed multiple tests, varying parameters such as the number of ants (ants), alpha, and beta. Each test was carefully designed to observe how changes in these parameters affect the algorithm's performance, specifically in terms of execution time, best distance (optimal route), and the number of vehicles used.

By plotting the data, we can clearly see how different combinations of parameters affect the computation time, best route length, and other important performance metrics. This approach allows us to fine-tune the algorithm and make informed decisions about which parameters provide the best performance for the given problem.

Our tests will be conducted on the C, R, and RC instances from the Solomon benchmark set. These instances are widely used in the vehicle routing problem (VRP) and provide different types of problem configurations to assess the robustness and efficiency of the algorithm under various conditions.

C Instances: These instances represent clustered customer locations. The customers are grouped together in clusters, making the problem more structured and often easier for routing algorithms to solve, as the vehicles typically serve customers in the same cluster.

R Instances: These instances represent random customer locations. The customers are distributed randomly across the geographical area, which typically results in a more challenging problem as the vehicles must cover a larger and more scattered set of locations.

RC Instances: These instances combine both clustered and random customer locations. Some customers are placed in clusters, while others are scattered randomly, providing a hybrid challenge for the routing algorithm to address.

By testing on these different types of instances, we can evaluate how the algorithm performs under various problem conditions, from well-structured clustered cases to more complex random and mixed distributions.
