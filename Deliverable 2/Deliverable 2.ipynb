## **IV. Modelisation<a class="anchor" id="m"></a>**

To start, we need to clarify the core need of the problem, which will guide our objectives and constraints.

**Need Identification:** A logistics provider needs to deliver goods to multiple customers efficiently. Efficient delivery means achieving optimal routes for multiple delivery vehicles to meet customer requirements while minimizing operational costs.

**Problem Type:** This problem can be formulated as a Vehicle Routing Problem (VRP) if multiple vehicles are used. If only one vehicle is used, it aligns with the Traveling Salesman Problem (TSP). In our case it's VRP.

### **1. Comprehensive Problem Representation<a class="anchor" id="cp"></a>**

#### **A. Why VRP and Not TSP?<a class="anchor" id="wp"></a>**
This problem is categorized as a **Vehicle Routing Problem (VRP)** rather than a **Traveling Salesman Problem (TSP)** because:  
- In the TSP, a single vehicle must visit all customers and return to the depot.  
- In contrast, our scenario involves **multiple vehicles**, requiring the optimization of customer allocation to vehicles and the determination of the best route for each vehicle.  

The VRP adds complexity compared to TSP, as it must address the assignment of customers to vehicles, while also considering factors like vehicle capacity and time windows.

##### **<u>Key Constraints</u>**

1. **Single Visit per Customer**
   - **Description:** Each customer must be visited exactly once, and only one vehicle can serve each customer.  
   - **Constraint Type:** Equality constraint.  
   - **Purpose:** Ensures that no customer is missed or served multiple times, avoiding redundant or overlapping deliveries.

2. **Depot Start and End**
   - **Description:** All vehicles must begin and end their routes at the depot (node 0).  
   - **Constraint Type:** Equality constraint.  
   - **Purpose:** Ensures proper route closure for operational consistency and planning.

3. **Capacity Constraint**
   - **Description:** Each vehicle has a maximum carrying capacity that cannot be exceeded by the total demand of its assigned customers.  
   - **Constraint Type:** Hard constraint.  
   - **Purpose:** Prevents vehicle overloading and ensures feasible route assignments.

4. **Time Windows**
   - **Description:** Deliveries must be made within a specified time window for each customer.  
   - **Constraint Type:** Hard or soft constraint (with penalties for violations).  
   - **Purpose:** Improves customer satisfaction and ensures realistic delivery schedules.

5. **Service Time**
   - **Description:** Time spent at each customer location for unloading or completing the service is included in the schedule.  
   - **Constraint Type:** Operational constraint.  
   - **Purpose:** Ensures realistic timing for routing and adherence to delivery schedules.

6. **Delivery-Leave Continuity**
   - **Description:** After serving a customer, a vehicle must leave either to visit another customer or to return to the depot.  
   - **Constraint Type:** Flow conservation constraint.  
   - **Purpose:** Ensures route continuity and prevents incomplete or disjointed paths.

#### **B. Mathematic Modelisation<a class="anchor" id="mm"></a>**
In this section, we present the mathematical modeling of the Vehicle Routing Problem (VRP), defining the necessary constraints and objectives for optimizing the routes. Each constraint is formulated using decision variables that model the relationships between customers, vehicles, and the depot, with the goal of minimizing the total cost while adhering to specific requirements such as capacity, time windows, and service times. Additionally, we define a fitness function, which serves to evaluate and optimize the solution, ensuring that the routes are both efficient and feasible according to the problem's constraints.

##### **<u>Decision Variables</u>**
In mathematical optimization and operations research, decision variables are the variables that represent the choices made in a problem. They are the unknowns that the optimization model seeks to determine or optimize. 

In the context of the Vehicle Routing Problem (VRP), decision variables define the routing choices, such as the paths taken by vehicles, the timing of deliveries, and the allocation of customers to vehicles. These variables are used in the objective function and constraints to describe the system's behavior and ensure that the solution satisfies all the necessary conditions.

Now, we define them:

1. **$x_{ij}^k$**
   - **Description:** Binary decision variable indicating whether vehicle $k$ travels from customer $i$ to customer $j$.
   - **Value:** 
     - $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$
     - $x_{ij}^k = 0$ otherwise

2. **$t_i^k$**
   - **Description:** Arrival time at customer $i$ for vehicle $k$.
   - **Value:** Continuous variable representing the time at which vehicle $k$ arrives at customer $i$.

3. **$T_k$**
   - **Description:** Binary decision variable indicating whether vehicle $k$ is used.
   - **Value:** 
     - $T_k = 1$ if vehicle $k$ is used
     - $T_k = 0$ if vehicle $k$ is not used

4. **$q_k$**
   - **Description:** Capacity of vehicle $k$.
   - **Value:** Continuous variable representing the maximum load that vehicle $k$ can carry.

5. **$d_i$**
   - **Description:** Demand of customer $i$.
   - **Value:** Continuous variable representing the quantity or demand of customer $i$ that must be met by a vehicle.

6. **$s_i$**
   - **Description:** Service time at customer $i$.
   - **Value:** Continuous variable representing the time required to service customer $i$ (e.g., unloading, acceptance).

##### **<u>Fitness Function for a Simple VRP</u>**


In a basic **Vehicle Routing Problem (VRP)**, the primary objective is to minimize the total distance traveled by all vehicles while ensuring that all customers are visited exactly once. The fitness function can be expressed as:

$$
\text{Minimize } Z_{\text{VRP}} = \sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \cdot x_{ij}^k
$$

**Where:**
- $c_{ij}$: The cost (e.g., distance or time) of traveling from node $i$ to node $j$.
- $x_{ij}^k$: A binary variable that equals 1 if vehicle $k$ travels from $i$ to $j$, and 0 otherwise.
- $V$: The set of all nodes (customers and depot).
- $K$: The set of available vehicles.

This fitness function simply minimizes the total travel cost, without considering additional constraints like vehicle capacity or customer-specific time windows.


##### **<u>From VRP to CVRPTW</u>**

While the simple VRP aims to minimize travel costs, real-world routing problems often include additional constraints, such as vehicle capacity and customer time windows. These extensions lead to the **Capacitated Vehicle Routing Problem with Time Windows (CVRPTW)**. The CVRPTW adds complexity by requiring that:

1. **Vehicle Capacities:** Each vehicle has a limited carrying capacity, and the total demand of customers assigned to a vehicle cannot exceed this capacity.
2. **Time Windows:** Each customer has a specific time window during which they must be serviced.

To incorporate these constraints, the fitness function must account for **multiple objectives**, such as minimizing the total cost while adhering to these constraints. Additionally, the model must ensure feasible solutions by enforcing capacity and time window constraints.



##### **<u>Fitness Function for the CVRPTW</u>**

The fitness function for the CVRPTW combines the objectives of minimizing the number of vehicles and the total travel cost, while ensuring feasible delivery times and capacity usage. A common formulation is:

$$
\text{Minimize } Z_{\text{CVRPTW}} = \left( \sum_{k \in K} T_k \right) \cdot \left( \sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \cdot x_{ij}^k + \sum_{i \in V} s_i \right)
$$

**Where:**
- $\sum_{k \in K} T_k$: Total number of vehicles used.
- $\sum_{k \in K} \sum_{i \in V} \sum_{j \in V} c_{ij} \cdot x_{ij}^k$: Total travel cost (e.g., distance, time, or fuel).
- $\sum_{i \in V} s_i$: Total service time for all customers.
- $x_{ij}^k$: Binary decision variable indicating whether vehicle $k$ travels from customer $i$ to customer $j$.
- $T_k$: Binary decision variable indicating whether vehicle $k$ is used.


##### **<u>Why Extend to the CVRPTW?</u>**

1. **Realistic Constraints:** In real-world applications, simple VRP models fail to capture operational challenges like limited vehicle capacity and customer time preferences. Adding these constraints makes the model more realistic and applicable to industries like logistics and supply chain management.
   
2. **Minimizing Total Costs:** By multiplying the number of vehicles used by the total travel cost, this fitness function penalizes solutions that use more vehicles or inefficient routing. It ensures a balanced trade-off between reducing the fleet size and optimizing routes.

3. **Feasibility:** Enforcing time windows and capacity ensures that the solution is not only optimal in terms of cost but also operationally feasible.

4. **Multi-Objective Nature:** The CVRPTW inherently involves multiple objectives (e.g., minimizing distance, adhering to time windows, and minimizing the fleet size). This fitness function integrates these objectives into a unified mathematical framework.


While the simple VRP focuses on minimizing travel costs, the CVRPTW addresses more complex, real-world challenges by incorporating capacity constraints and time windows. The fitness function evolves to balance multiple objectives, ensuring feasible and cost-effective routing solutions. By integrating additional constraints into the fitness function, the CVRPTW model becomes a powerful tool for solving practical logistics and delivery problems.



##### **<u>Constraints</u>**

##### **1. <u>Single Visit per Customer Constraint</u>**

**Description:**  
Each customer must be visited exactly once by a single vehicle. No customer can be omitted or visited multiple times.

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.

**Model:**  
$$
\sum_{k=1}^K \sum_{j=1, j \neq i}^N x_{ij}^k = 1 \quad \forall i \in \{1, \dots, N\}
$$  
This ensures that each customer is visited exactly once.

**Explanation:**  
This constraint ensures that each customer is visited only once by a vehicle, and no customer will be missed or visited multiple times.

---

##### **2. <u>Depot Start and End Constraint</u>**

**Description:**  
Each vehicle must start and end its route at the depot (node 0).

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.

**Model:**  
$$
\sum_{j=1}^N x_{0j}^k = 1 \quad \forall k \in \{1, \dots, K\} \quad \text{(Vehicle starts from the depot)}
$$  
$$
\sum_{i=1}^N x_{i0}^k = 1 \quad \forall k \in \{1, \dots, K\} \quad \text{(Vehicle returns to the depot)}
$$

**Explanation:**  
This constraint ensures that each vehicle starts and ends its route at the depot, thus completing the delivery loop.

---

##### **3. <u>Delivery-Leave Continuity</u>**

**Description:**  
When a vehicle arrives at a customer, it must either proceed to another customer or return to the depot. This ensures the continuity of the route without any breaks.

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.  
- $t_i^k$: Arrival time at customer $i$ for vehicle $k$.

**Model:**  
1. **Inflow and Outflow for each customer $i$ and vehicle $k$:**
   $$
   \sum_{j \neq i} x_{ij}^k = \sum_{j \neq i} x_{ji}^k \quad \forall i \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
   $$  
   This ensures that a vehicle arrives at each customer exactly once and departs exactly once.

**Explanation:**  
This constraint ensures that vehicles continue their routes without interruption.

---

##### **4. <u>Time Window Constraint</u>**

**Description:**  
Each customer has a specific time window during which they wish to receive their delivery. The delivery must occur within this window, or with a penalty if the constraint is softened.

**Decision Variable:**  
- $t_i^k$: Arrival time at customer $i$ for vehicle $k$.

**Model:**  
1. **Arrival Time:**
$$
a_i \leq t_i^k \leq b_i \quad \forall i \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
$$  
Where $a_i$ and $b_i$ are the time window bounds for customer $i$.

2. **Arrival time calculation $t_i^k$:**
   $$
   t_j^k \geq t_i^k + s_i + d_{ij} - M(1 - x_{ij}^k) \quad \forall i, j \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
   $$  
   Where $M$ is a large constant to ensure the constraint is inactive if $x_{ij}^k = 0$.

**Explanation:**  
This constraint ensures that each delivery respects the time window defined for each customer. If a time window is exceeded, a penalty could be added to the model. And that the arrival time at each customer is calculated based on the previous route.

---

##### **5. <u>Capacity and Demand Constraint</u>**

**Description:**  
Each vehicle has a maximum capacity, and each customer has a specific demand that must be met without exceeding this capacity.

**Decision Variable:**  
- $x_{ij}^k = 1$ if vehicle $k$ travels from $i$ to $j$, otherwise $0$.  
- $q_k$: Capacity of vehicle $k$.  
- $d_i$: Demand of customer $i$.

**Model:**  
$$
\sum_{i=1}^N \sum_{j=1, j \neq i}^N d_i \cdot x_{ij}^k \leq q_k \quad \forall k \in \{1, \dots, K\}
$$

**Explanation:**  
This constraint ensures that the demand of each customer is met while respecting the capacity of each vehicle. The sum of the demands for all customers visited by a vehicle must not exceed the capacity of the vehicle.

---

##### **6. <u>Service Time Constraint</u>**

**Description:**  
The service time represents the time required for a vehicle to complete a delivery at a customer, including unloading time or waiting for acceptance.

**Decision Variable:**  
- $t_i^k$: Arrival time at customer $i$ for vehicle $k$.  
- $s_i$: Service time at customer $i$.

**Model:**  
$$
t_j^k \geq t_i^k + s_i + d_{ij} - M(1 - x_{ij}^k) \quad \forall i, j \in \{1, \dots, N\}, \, \forall k \in \{1, \dots, K\}
$$

**Explanation:**  
This constraint accounts for the service time at each customer location, ensuring that the scheduling of routes and timings is realistic.

---

### **2. Metaheuristic Techniques and Formal Modelling Detail<a class="anchor" id="mt"></a>**

#### **A. Complexity<a class="anchor" id="cx"></a>**

##### **<u>VRPTW and its Relation to VRP</u>**

VRPTW is an extension of VRP. In VRPTW, each customer has a specific time window during which they must be visited, in addition to vehicle capacity constraints.  
To reduce VRPTW to VRP, we can consider a simplified case where time windows are the same as customer schedules, so they impose no constraints on delivery times. This simplification removes the time window constraints, reducing VRPTW to a standard case of VRP. Consequently, any algorithm that solves VRPTW can also solve VRP as a special case, indicating that VRPTW is at least as difficult as VRP.

##### **<u>Proof that VRP is NP-Complete</u>**

##### **1. <u>VRP is in NP</u>**

To establish that the Vehicle Routing Problem (VRP) is in the class NP, we need to demonstrate that, given a candidate solution, we can verify its correctness in polynomial time. Specifically, we consider the Decision Version of VRP, where we ask whether there exists a set of routes for the vehicles such that all customers are served, all constraints are satisfied, and the total cost does not exceed a given value \(k\).

Given a proposed set of routes for the vehicles, the verification process involves the following checks, each of which can be performed in polynomial time:

- **a. Each customer is visited exactly once**  
  - Verification: Ensure that every customer appears in exactly one route.  
  - Time Complexity: Linear time $O(n)$, where $n$ is the number of customers.

- **b. Vehicle Capacity Constraints are Respected**  
  - Verification: For each vehicle, sum the demands of the customers on its route and ensure it does not exceed the vehicle’s capacity.  
  - Time Complexity: Linear time $O(n)$, as it involves iterating through each route and summing demands.

- **c. Routes Start and End at the Depot**  
  - Verification: Check that each route begins and ends at the designated depot.  
  - Time Complexity: Linear time $O(m)$, where $m$ is the number of vehicles (typically $m \leq n$).

- **d. Total Cost Does Not Exceed $k$**  
  - Verification: Calculate the total cost (e.g., total distance) by summing the costs of all routes and verify it is less than or equal to $k$.  
  - Time Complexity: Linear time $O(n)$, as it involves summing the costs of all routes.

Since all these verifications can be performed in linear time relative to the size of the input, VRP is in NP.

##### **2. <u>Reduction from Hamiltonian Cycle to VRPTW</u>**

To prove that VRP is NP-Hard, we will reduce a known NP-Complete problem to VRP in polynomial time. We choose the Hamiltonian Cycle Problem for this reduction.

**Hamiltonian Cycle Problem**  
- Input: An undirected graph $G = (V, E)$.  
- Question: Does there exist a cycle that visits each vertex exactly once and returns to the starting vertex?


**1. Reduction Process:**

- **Instance Transformation:**  
  - Given: An undirected graph $G = (V, E)$ for the Hamiltonian Cycle problem.  
  - Construct: A VRP instance as follows:  
    - **Depot:** Choose an arbitrary vertex $D$ from $V$ as the depot.  
    - **Customers:** All other vertices $V' = V \setminus D$ are treated as customers, each with a demand of 1.  
    - **Vehicle Capacity:** Set the vehicle capacity $C = |V|$, ensuring a single vehicle can serve all customers.  
    - **Distances:** Define the distance between two vertices $u$ and $v$ as:  
    $$
    d(u, v) = 
    \begin{cases} 
      1 & \text{if } (u, v) \in E \\
      2 & \text{if } (u, v) \notin E 
    \end{cases}
    $$
    - **Total Cost $k$:** Set $k = |V|$.

**2. Equivalence of Solutions:**  
- If $G$ has a Hamiltonian Cycle:  
  The corresponding VRPTW instance can be solved by having the single vehicle follow the Hamiltonian cycle, resulting in a total cost of $|V|$ (since each edge in the cycle has a distance of 1).
- If the VRP instance has a solution with total cost $\leq k$:  
  The vehicle’s route must use only edges with distance 1 to keep the total cost at $|V|$. This implies the existence of a Hamiltonian Cycle in $G$.

**3. Polynomial-Time Transformation:**  
- The reduction involves creating a VRP instance with:  
  - $|V|$ customers.  
  - A complete graph for distances, requiring $O(|V|^2)$ time.  
  - Setting capacities and costs, which are linear operations.  
- Therefore, the reduction is performed in polynomial time.

#### **B. Metaheuristic Algorithm<a class="anchor" id="ma"></a>**

#### **Heuristic Algorithms**
**Definition:** Heuristic algorithms are problem-specific approaches designed to provide fast and feasible solutions. They rely on rules or strategies derived from the structure of the problem.
**How They Work:** Heuristics use simple strategies like greedy selection (e.g., selecting the next best option) or local adjustments to improve solutions. For example, in routing problems, the Nearest Neighbor Algorithm assigns routes by selecting the closest available customer.
##### **Advantages:**
Quick and computationally cheap.
Simple to implement.
##### **Disadvantages:**
Often stuck in local optima (poor-quality solutions).
Cannot effectively handle large, complex problems.
#### **Metaheuristic Algorithms**
**Definition:** Metaheuristics are higher-level, general frameworks that guide the search process to explore a problem's solution space more broadly and deeply than heuristics.
##### **How They Work:**
Metaheuristics are not tied to specific problems; they can be adapted for various optimization tasks.
They use mechanisms such as randomness and iterative improvement to avoid getting stuck in local optima.
##### **Examples include:**
**Genetic Algorithms (GA):** Simulates evolution via selection, crossover, and mutation.
**Simulated Annealing (SA):** Mimics physical annealing to escape local minima.
**Ant Colony Optimization (ACO):** Simulates ants’ pheromone trails to optimize paths.
##### **Advantages:**
Capable of finding near-optimal solutions.
Effective for large, complex, or non-linear problems like CVRPTW.
##### **Disadvantages:**
Computationally more expensive than heuristics.
Requires careful parameter tuning.
#### **Hybrid Algorithms**
**Definition:** Hybrid algorithms combine the strengths of heuristic and metaheuristic techniques, or different metaheuristics, to leverage their advantages and improve solution quality.
##### **How They Work:**
For instance, a hybrid algorithm may use a genetic algorithm for global exploration and a local search heuristic for fine-tuning.
These approaches are tailored for specific problems to achieve both scalability and quality.
##### **Advantages:**
Better performance on real-world, complex problems.
Combines global search (metaheuristic) with fast local improvements (heuristic).
##### **Disadvantages:**
Higher computational and implementation complexity.
Why Use Metaheuristics for CVRPTW?
The Capacitated Vehicle Routing Problem with Time Windows (CVRPTW) is a complex combinatorial optimization problem where:

Vehicles must serve customers within predefined time windows.
Routes must minimize costs (e.g., distance or time) while respecting vehicle capacity constraints.
##### **Challenges of CVRPTW:**
Large Search Space: The number of possible routes grows exponentially with the number of customers.
##### **Conflicting Constraints:**

##### **Metaheuristics Solve CVRPTW Effectively:**
**Escape from Local Optima:** Metaheuristics explore the solution space broadly, avoiding suboptimal solutions where traditional heuristics often get stuck.
**Scalability:** Metaheuristics can handle the large-scale complexity of CVRPTW with hundreds or thousands of customers.

##### **<u>Introduction to Metaheuristic Algorithms</u>**

Metaheuristic algorithms are problem-solving methods designed to find approximate solutions to complex optimization problems, especially those that are NP-hard. These algorithms are useful when the solution space is large or non-linear, like in the **Vehicle Routing Problem with Time Windows (VRPTW)**. Metaheuristics explore vast search spaces to find good-enough solutions in a reasonable time.

They are inspired by natural processes and handle large, complex tasks efficiently. While they do not guarantee optimal solutions, they typically find close-to-optimal solutions quickly.

##### **<u>Key Features of Metaheuristic Algorithms:</u>**
- **Exploration and Exploitation**: Balances searching broad areas (exploration) and refining promising areas (exploitation).
- **Stochastic Nature**: Randomness helps avoid getting stuck in local optima.
- **Adaptability**: Can be customized for various problems by modifying components like solution representation and operators (e.g., crossover, mutation).
- **Flexibility**: Can solve a wide range of combinatorial optimization problems, including VRPTW.


##### **1. Algorithm Choice and Description:**

<b>Genetic Algorithm :  a class of metaheuristic optimization algorithms inspired by the process of natural selection and the principles of genetics. They are widely used to solve complex problems where traditional optimization methods might struggle.</b>

  1. Representation (Encoding)
Chromosomes: In GAs, potential solutions are represented as chromosomes, often encoded as strings of binary digits (0s and 1s), real numbers, or other forms, depending on the problem. These chromosomes represent a candidate solution to the problem.
  
  2. Initial Population
GAs start with a population of randomly generated solutions. The size of the population is a key parameter and can influence the performance of the algorithm.
  
  3. Fitness Function
The fitness function is used to evaluate how good each solution (chromosome) is with respect to the problem. The fitness value determines how likely a solution will be selected for reproduction. The better the fitness, the higher the chance of selection.

  4. Selection
In the selection phase, solutions (chromosomes) are chosen based on their fitness to create offspring. Common selection methods include roulette wheel selection, tournament selection, and rank-based selection. Solutions with higher fitness have a greater chance of being selected.
  
  5. Crossover (Recombination)
After selection, two parent solutions are paired and undergo crossover, a process where parts of their genetic material (chromosomes) are exchanged to create one or more offspring. This simulates biological recombination and introduces diversity into the population.
  
  6. Mutation
After crossover, mutation introduces random changes to the offspring. This is done by flipping bits or modifying values at random positions in the chromosome. Mutation ensures diversity within the population and prevents the algorithm from getting stuck in local optima.
  
  7. Replacement
The offspring created by crossover and mutation replace some or all of the old population, depending on the replacement strategy. The new population is then evaluated, and the process repeats until a termination condition is met (e.g., a certain number of generations or a satisfactory solution).



<b>Ant Colony Optimization : is a nature-inspired optimization algorithm based on the foraging behavior of ants. It is particularly effective for solving combinatorial optimization problems, such as the Traveling Salesman Problem (TSP), Vehicle Routing Problem (VRP), and Quadratic Assignment Problem (QAP). ACO mimics how real ants find the shortest path between their colony and food sources, with the help of pheromone trails.</b>


  1. Pheromone Representation
Pheromone Trails: In ACO, ants use pheromones to communicate indirectly with each other. Pheromones are a form of chemical markers that ants leave on their paths. The strength of the pheromone trail influences the decisions of other ants to follow the same path.

  2. Ant Movement (Solution Construction)
Each ant constructs a solution (or path) by moving from one node to another, making decisions based on two key factors:
Pheromone concentration: Ants are more likely to follow paths with stronger pheromone concentrations, which are considered more promising based on previous ants' experiences.
Heuristic information: Some problems may include additional information (like distance or cost) that influences an ant’s path selection.

  3.Exploration vs. Exploitation
Exploitation: Ants prefer paths with high pheromone levels, which means they exploit the best paths discovered so far.
Exploration: Ants sometimes choose less-pheroed paths with some probability, promoting exploration of new paths and preventing premature convergence on suboptimal solutions.

#### **Ant Colony Optimization for VRPTW**
Initialization of the Colony: The first step in ACO is to initialize the environment and parameters for the colony.

Graph Representation:
The problem is modeled as a graph where nodes represent customers (and the depot).
Edges between nodes represent the distances between customers.
Pheromone Trails:
A pheromone matrix is initialized with a constant value. This matrix helps ants decide which routes to explore based on accumulated pheromone levels.
Heuristic Information:
A heuristic factor is calculated for each edge. For example, the inverse of the distance between two nodes encourages ants to favor shorter routes.
Parameters:
Define ACO-specific parameters like:
Number of Ants: Number of solutions (ants) generated per iteration.
Alpha: Weight of pheromone importance.
Beta: Weight of heuristic importance.
Evaporation Rate: Controls the rate at which pheromone trails fade.
Pheromone Deposit: Amount of pheromone deposited after an ant completes its route.
Ant Tour Construction: Each ant builds a feasible solution (route) by visiting customers sequentially, adhering to VRPTW constraints.

Steps:

Starting at the Depot:
All ants begin their tours at the depot (node 0).
Next Node Selection:
For each customer, an ant chooses the next node to visit based on:
Pheromone Level: Edges with more pheromones are more likely to be chosen.
Heuristic Value: Edges with shorter distances or better time feasibility are preferred.
Feasibility Check: The chosen node must satisfy:
Vehicle Capacity: Demand of the customer cannot exceed the remaining capacity.
Time Windows: Arrival time must fall within the customer’s ready and due times.
Move to Node:
The ant updates its current position, route, remaining capacity, and current time.
Completion:

The ant’s route ends when all customers are visited, after which it returns to the depot.
Pheromone Update: After all ants in the colony complete their tours:
Evaporation:
Pheromones on all edges decay by a factor (controlled by the evaporation rate). This prevents early convergence by gradually reducing the influence of previous paths.
Deposit:
Pheromones are deposited on the edges of the best-performing routes (or all routes, depending on the strategy). The amount of pheromone deposited is inversely proportional to the total cost of the route (distance or time penalties).
Fitness Function: ACO evaluates the quality of each solution (route) using a fitness function:

Route Distance:
Minimize the total distance traveled by all vehicles.
Time Window Constraints:
Penalize routes where customers are served too late or too early.
Capacity Constraints:
Penalize routes that exceed vehicle capacity.
Objective: Minimize total cost = Distance + Penalty for time violations + Penalty for capacity violations.

Iterative Process: The ACO algorithm iteratively improves the solutions by balancing exploration (searching new routes) and exploitation (refining good routes).

Steps:

Ants Construct Routes:
Each ant builds a feasible route using pheromone and heuristic information.
Evaluate Solutions:
Compute the fitness of each route.
Update Pheromones:
Adjust the pheromone matrix based on the quality of the routes.
Update Best Solution:
If an ant produces a route better than the current best solution, update the best solution.
Termination:

The process repeats for a fixed number of iterations or until the best solution stabilizes.
Final Solution: After the algorithm terminates, the best route found by the ants is returned as the solution.


#### **Genetic algorithm**

The goal of this genetic algorithm is to solve the Vehicle Routing Problem with Time Windows (VRPTW), where a fleet of vehicles must deliver goods to a set of customers while satisfying the following constraints:

Each customer has a specific time window in which they need to be served.
Each vehicle has a fixed capacity and can only handle a certain amount of demand.
Each route must start and end at the same node.

1. Initial Population Generation:
The first step of the genetic algorithm is to generate an initial population of potential solutions (routes).

Each individual (solution) in the population is represented as a list of customer indices, with the depot implicitly assumed to be the starting and ending point of the routes.

Functionality: Randomly shuffle the customers to generate different routes for each vehicle.

2. Fitness Function:
The fitness function evaluates the quality of a solution. In this case, the goal is to minimize the total distance of the routes and penalize violations of the time window constraints.

Steps:

For each route, the algorithm calculates the total distance traveled.
It then checks if each customer is served within their time window. If a customer is served too early, it incurs a penalty for waiting. If they are served too late, a higher penalty is added for violating the due date.
Fitness Objective: The fitness function penalizes both the distance of the route and any violations of the time windows. The goal is to minimize both.


3. Selection (Tournament Selection):
Tournament Selection is used to select the best parents for the next generation. In tournament selection, a subset of individuals (a "tournament") is selected randomly, and the best individual from that tournament is chosen.

Steps:

        a)A random subset of solutions (tournament size) is selected.


        b)The solutions are sorted by their fitness, and the top two individuals are chosen as parents.


4. Crossover:
Crossover combines the genetic information from two parent solutions to create offspring. It selects segments of each parent’s route and combines them to form new routes for the children.

5. Mutation:
The mutation operation introduces small random changes to a solution. This helps maintain diversity in the population and prevents the algorithm from prematurely converging.

6. Main Evolution Loop:
The algorithm iterates through multiple generations to evolve the population. In each generation:
                
                Parents are selected.
                
                Crossover and mutation are applied to generate offspring.
                
                The population is replaced with the new generation.
                
                The best solution is updated based on fitness.
